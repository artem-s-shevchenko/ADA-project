# Visualisation

```python
from operator import itemgetter
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data=[]
f = open('visual_brand_mean.txt', 'r')
for line in f: data.append(line)
data[0:4]
```

```python
brands_names=[]
brands_means=[]
for x in range(0,len(data),2): brands_names.append(data[x][0:-1])
for x in range(1,len(data),2): brands_means.append(float(data[x][0:-1]))
pairs = [(x,y) for x,y in zip(brands_names,brands_means)]
pairs = sorted(pairs, key=itemgetter(1))
brands_names,brands_means = zip(*pairs)
X=range(0,len(brands_names),3)
brands_means=brands_means[0::3]
plt.figure(figsize = (20, 10))
x_span = max(X)-min(X)
cm = plt.cm.get_cmap('RdYlBu_r')
C = [cm(((x-min(X))/x_span)) for x in range(248,-1,-3)]
plt.bar(X, brands_means, color=C, align='center', width=3)
plt.xticks(X, brands_names[0::3])
```
